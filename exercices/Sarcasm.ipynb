# Sarcasm Text Analysis

```python
import json
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

## Importation of data

```python
# with open('./../data/Sarcasm_Headlines_Dataset.json', 'r') as f:
#     datastore = json.load(f)
datastore = [json.loads(line) for line in open('./../data/Sarcasm_Headlines_Dataset.json', 'r')]
```

```python
sentences = []
labels = []
urls = []

for item in datastore:
    sentences.append(item['headline'])
    labels.append(item['is_sarcastic'])
    urls.append(item['article_link'])
```

```python

tokonizer = Tokenizer(oov_token="<OOV>")
tokonizer.fit_on_texts(sentences)
word_index = tokonizer.word_index
```

```python
sequences = tokonizer.texts_to_sequences(sentences)
sequences
```

```python
padded = pad_sequences(sequences)
```
